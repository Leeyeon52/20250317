```python
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df = pd.read_csv('data/kr_population.csv',index_col=0, parse_dates=True)
```

```python
df[:15]
```

```python
df['birth'].rolling(window=10).mean()[:15]
```

```python
pit.style.use('grayscale')
pit.figure(facecolor='white')

df['birth'].plot(label='Num of prople born in s.korea',linestyle='-')

df['birth'].rolling(window=10).mean().plot(label='Rolling Mean(window=10)',linestyle=':')

df['birth'].expanding().mean().plot(label='Expanding', linestyle='--')

pit.legend()
```

```python
pit.style.use('grayscale')
pit.Figure(facecolor='white')

df['birth'].rolling(window=10).sum().plot(label='Rolling Sum(window=10)',linestyle=':')

df['birth'].expanding().sum().plot(label='Expanding',linestyle='-.')
pit.legend()
```

```python
pit.style.use('grayscale')
pit.figure(facecolor='white')

df['birthrate'].plot(label='Population growth Rate', linestyle='-')

df['birthrate'].rolling(window=10).mean().plot(label='Rolling Mean(Window=10)',linestyle=':')

df['birthrate'].expanding().mean().plot(label='Expanding',linestyle='--')
pit.legend()
```

```python
pit.style.use('grayscale')
pit.Figure(facecolor='white')

df['birthrate'].fillna(method='ffill').plot(label='Population growth Rate', linestyle='-')

df['birthrate'].fillna(method='ffill').rolling(window=10).mean().plot(label='Rolling Mean(window10)',linestyle=':')

df['birthrate'].fillna(method='ffill').expanding().mean().plot(label='Expandig', linestyle='--')

pit.axhline(y=2.0, xmin=0.6, xmax=1.0, color='lightgray',linestyle='--', linewidth=2)
pit.axhline(y=2.0, xmin=0.6, xmax=1.0, color='lightgray',linestyle='-', linewidth=2)

pit.legend()
pit.show()
```
